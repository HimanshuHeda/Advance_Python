# Animation

```python
import matplotlib.animation as animation
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML

fig,ax=plt.subplots()
ax.set_xlim(0,150)
ax.set_ylim(0,10)

x=np.linspace(0,150,100)
y=np.linspace(0,10,100)

point, = ax.plot([],[],'o',markersize=10)

def funcani(frame):
    point.set_data([x[frame]],[y[frame]])
    return point,

a=animation.FuncAnimation(fig,funcani,frames=np.arange(0,100,1),interval=20,blit=False)

# plt.close(fig)

HTML(a.to_jshtml())
```
