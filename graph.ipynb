# Bar Charts - 2D Bar Charts

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./marksheet.csv")
```

```python
plt.bar(data['Name'],data['Science'])
```

```python
plt.xticks(rotation=90)
plt.bar
```

# Stacked Bar Chart

```python
plt.figure(figsize=(10,8))
plt.bar(data['Name'], data['Science'])
plt.bar(data['Name'], data['English'], bottom=data['Science'])
plt.bar(data['Name'], data['History'], bottom=data['Science'] +data['English'])
plt.bar(data['Name'], data['Maths'], bottom=data['Science'] +data['English'] +data['History'])

plt.xticks(rotation=90)

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("2D Bar Graph")
```

```python
subjects = ['Science', 'English', 'History', 'Maths']

fig, ax = plt.subplots(figsize=(15,8))
x = np.arange(len(data['Name']))
width = 0.2

for i, subject in enumerate(subjects):
    plt.bar(x + i * width, data[subject], width, label=subject)
```

```python
subjects = ["Science", "English", "History", "Maths"]

x = np.arange(len(data["Name"]))
width = 0.2

fig, ax = plt.subplots(figsize=(15,6))

for i,subject in enumerate(subjects):
    ax.bar(x + i * width, data[subject], width, label=subject)

ax.set_xlabel("Students")
ax.set_ylabel("Scores")
ax.set_title("Students' Scores in DIfferent Subjects")
ax.set_xticks(x + width)
ax.set_xticklabels(data["Name"], rotation=90)
ax.legend(title="Subjects")

plt.tight_layout()
plt.show()
```

# Scatter Plot

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(data['Science'], data['English'], data['History'], color='blue', alpha=0.3)

for i in range(len(data['id'])):
    ax.text(data['Science'].values[i], data['English'].values[i], data['History'].values[i], data['Name'].values[i])
plt.show()
```
