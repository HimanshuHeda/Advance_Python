Bar Charts - 2D Bar Charts

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("./marksheet.csv")
```

```python
plt.bar(data['Name'],data['Science'])
```

```python
plt.xticks(rotation=90)
plt.bar
```

Stacked Bar Chart

```python
plt.figure(figsize=(10,8))
plt.bar(data['Name'], data['Science'])
plt.bar(data['Name'], data['English'], bottom=data['Science'])
plt.bar(data['Name'], data['History'], bottom=data['Science'] +data['English'])
plt.bar(data['Name'], data['Maths'], bottom=data['Science'] +data['English'] +data['History'])

plt.xticks(rotation=90)

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("2D Bar Graph")
```
