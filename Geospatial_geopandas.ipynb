# Importing Libraries

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

# Use to import Shape File

```python
world = gpd.read_file("./geopandas/ne_110m_admin_0_countries.shp")
```

# It is use to Read the Data 

```python
world.head()
```

# It is use to display the World

```python
world.plot()
```

# It is use to change the color and border color 

```python
fig, ax = plt.subplots(figsize=(15, 15))

world.plot(ax=ax, color="lightgreen", edgecolor="black")
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

world.plot(ax=ax, cmap="coolwarm", edgecolor="black")
```

# New India Map is Imported 

```python
india = gpd.read_file("./India/Indian_States.shp")
```

```python
india.head()
```

```python
india.info()
```

```python
india.plot()
```

```python
india.plot( cmap="Pastel1", edgecolor="black")

plt.title("Indian - States Map", fontsize=14)
```

```python
india["st_nm"] = india["st_nm"].str.strip()

state = india[india["st_nm"] == "Assam"]

ax = india.plot(color="lightgreen", edgecolor="black", figsize=(8, 6))

state.plot(ax=ax, color="red")

plt.title("Highlighting Assam", fontsize=14)
```
